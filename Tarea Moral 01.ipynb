** 1.  Crea un script que calcule los primeros N números de fibonacci**

```python
def fibo (n):
    a = list(range(n))
    for i in range(n):
        if i==0:
            a[i]=0
        elif i==1:
            a[i]=1
        else:
            a[i]=a[i-2] + a[i-1]
    return a
    
```

```python
fibo(15)
```

** 2. Crea una función que calcule los coeficientes de binomiales $C(n,k)$ **

```python
def cbinom(n,k):
    import math as m
    if (k <=n) & (k>=0):
        return  int(m.factorial(n) / (m.factorial(k)*m.factorial(n-k)))
```

```python
cbinom(4,3)
```

```python
def cbinomInt(n,k):
    if k==0:
        return 1
    if k==n:
        return 1
    else:
        return cbinomInt(n-1,k-1) + cbinomInt(n-1,k)
```

```python
cbinomInt(4,3)
```
