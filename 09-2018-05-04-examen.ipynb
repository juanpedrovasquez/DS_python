```python
import numpy as np
import pandas as pd
accidentes = pd.read_csv('data/accidentes.csv')
accidentes.head()
```

```python
dos=accidentes.loc[:,['Category','AccidentID']].groupby('Category').agg('count')
dos
```

```python
import geopandas as gpd
```

```python
municipios = gpd.read_file('data/Municipios/')
municipios.head()
```

```python
municipios.loc[municipios.NOM_MUN=='Xochimilco',:]
```

```python
cdmx = municipios.loc[municipios.CVE_ENT=='09',['NOM_MUN','geometry']]
cdmx.head()
```

```python
cdmx.crs
```

```python
cdmx=cdmx.to_crs( {'init': 'epsg:3395'} )
cdmx.head()
```

```python
from shapely.geometry import Point, Polygon, LineString
l = [Point(xy) for xy in zip(accidentes.Longitude , accidentes.Latitude)]
l[0:5]
```

```python
accidentes_gpd = gpd.GeoDataFrame(accidentes, geometry=l)
accidentes_gpd.head()
```

```python
accidentes_gpd = accidentes_gpd.fillna(0)
```

```python
accidentes_gpd.crs={'init':'epsg:4326'}
print(accidentes_gpd.crs)
accidentes_gpd.head()
```

```python
accidentes_gpd=accidentes_gpd.to_crs({'init': 'epsg:3395'})
accidentes_gpd.head()
```

```python
res_intersect = gpd.sjoin(cdmx,accidentes_gpd, how='left', op='intersects')
res_intersect.head()
```

```python
final = res_intersect.loc[:,['NOM_MUN','Category']].groupby('NOM_MUN').agg('count')
final.head()
```
